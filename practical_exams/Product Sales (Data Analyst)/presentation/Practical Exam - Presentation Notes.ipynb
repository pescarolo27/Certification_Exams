Before diving in, establishing some background information can help set the stage. Since 1984, the company _Pens & Printers_ has provided high quality office products to customers to aid their creativity & ability to brainstorm.

As the ways in which consumers buy products changes, sales tactics must adapt as necessary. As you may know, the company launched a new product line about six weeks ago which is an expensive process. As such, the sales team needs to make sure they are using the best techniques to sell the new products effectively. The best approach may vary for each new product so the team needs to learn quickly what works & what doesn't.

3 different sales strategies were tested for these purposes.
- There's the **"Email"** method in which customers received only an email. This required very little work for the team.
- There's the **"Call"** method in which customers received only a phone call from the sales team. On average, these calls lasted around 30 minutes per customer.
- Then there's the **"Email & Call"** method in which customers were contacted via both methods. These calls lasted about 10 minutes per customer.

## Business Metrics (1) (read here)
- estimated time: 70s --> 50-55s

So far, the analyses have come to multiple conclusions as to which sales method was the most effective. To help evaluate this further, a revenue efficiency metric was built that is based on the revenue generated & time spent advertising to a customer.  
For example, the typical revenue per customer in the "Email" group divided by the typical time spent advertising to these customers gives an estimate of the efficiency regarding how much revenue is generated over a period of time that is spent advertising via email.

For this metric, the average time spent contacting customers for each sales approach needs to be defined.
- Recall that for the **"Call"** approach, the calls lasted about 30 minutes.
- For the **"Email"** method, it is estimated that constructing & sending an email took no more 3 minutes on average.
- For customers contacted by **both email & phone**, the calls lasted about 10 minutes. So adding the email would total about 13 minutes on average.

## Recommendations (1) (read here)
- estimated time: 55s --> 50-55s

In combining the findings from all analyses, it's clear that the "Call" method was the least productive sales strategy by far in terms of both outright revenue as well as revenue efficiency. Ideally, this technique, given how much less profitable it has been in comparison to the other two methods, should be utilized at a minimum; however, this is not to say that talking to customers directly over the phone does not have some value.
- Some customers may prefer this kind of communication.

Moreover, with how long these phone calls typically take, it would be more productive for them to spend most of their time emailing customers & occasionally calling some people some time after emailing them.


Utilizing the "Email + Call" method resulted in the most revenue per customer, but this method hasn't been quite as efficient as only emailing people. Therefore, the "Email" method should be the primary sales approach, which hasn't been the case in the last two weeks.
