### Importing, Preprocessing, Exploratory Data Analysis

To begin, the data needs to be imported & inspected to determine if any preprocessing is required prior to analysis.

```python
#Import the data
orig_df = pd.read_csv("product_sales.csv")

#Inspect the data
display(orig_df.head(5))
display(orig_df.info())
```

There are 15,000 observations & eight variables in the original dataset. One variable ('revenue') has missing values (1,074). The data types for each variable seem appropriate. Next, the values in each variable can be perused to look for strange or unusual values that may point to typos, entry errors, or some other inconsistency.

```python
#Inspect values across columns
#display(orig_df['state'].value_counts(dropna=False).sort_values())
#display(orig_df['state'].nunique())

#Inspect data points w/missing 'revenue' values
display(orig_df[orig_df.revenue.isna()]['sales_method'].value_counts(dropna=False))
```

A summary of the values of each variable is listed below. Notes are included as to whether any variable needs preprocessing before analysis.
- **week:** Values span 1-6 (six unique values), each of which corresponds to between 1,228-3,721 data points. In other words, products have been sold between one & six weeks since their launch.
    - Recall that a new line of office stationery was launched six weeks prior to this dataset being assembled & sent to the Analytics team.
- **sales_method:** There are five unique values but some overlap. There are three general groups. For example, the values "Email + Call" & "em + call" can be grouped together as can "Email" & "email." Each group has between 2,500-7,500 data points.
- **customer_id:** There are 15,000 unique values. Therefore, there is only a single data point for each customer.
- **nb_sold:** Values span 7-16 (ten unique values), each of which corresponds to between 31-3,677 data points. In other words, the 15,000 customers purchased between seven & 16 items over the last six weeks.
- **revenue:** Values span 32.54-238.32, but there are 1,074 missing values. The units of this variable are unclear, but they will presumed to be US dollars ($). Perhaps the missing values could be imputed with some statistical value of this variable like the mean or median. Moreover, it is worth investigating if these missing values correspond to specific data points to determine if they are missing completely at random or not. This could influence the manner in which the missing values are handled.
- **years_as_customer:** Values span 0-63 years (42 unique values), each of which corresponds to between one & 2,504 data points. Currently, it is near the end of 2025. Given that the company was founded in 1984, this would indicate that customers could have been making purchases from _Pens & Printers_ for no more than 42 years (including 1984). There are two values in this variable that exceed this; 47 & 63.
    - If a customer has made purchases from the company for 47 or 63 years, it would indicate that the sales data is from, at the very least, 2030 or 2046 respectively. Such a large time gap does not seem sensible. As such, these two "insensible" values could be truncated to the realistic maximum number of years (42).
- **nb_site_visits:** Values span 12-41 (27 unique values), each of which corresponds to between one & 1,688 data points. In other words, customers have visited the company website between one & 1,688 times in the last six months.
- **state:** All 50 US states are accounted for. Between 28-1,872 orders have been shipped to each state.

In summary, the variables 'sales_method,' 'revenue,' & 'years_as_customer' need some preprocessing done.

```python
#Make a copy of the data to avoid modifying the original
df = orig_df.copy()

#Clean the 'sales_method' variable. Alter some values
df['sales_method'] = df['sales_method'].replace({'em + call':'Email + Call', 'email':'Email'})

#Clean the 'years_as_customer' variable. Truncate some values
df['years_as_customer'] = df['years_as_customer'].replace({47:42, 63:42})


#Evaluate the 'revenue' variable to determine how to handle the missing values.
    #Filter for data points with missing 'revenue's, inspect other variables to search for potential relationships
missing_df = orig_df[orig_df['revenue'].isna()]
display(orig_df['week'].value_counts(dropna=False).sort_index())
display(missing_df['week'].value_counts(dropna=False).sort_index())
```

The values of each variable were analyzed with respect to the data points with missing 'revenue' values to search for a potential relationship. For example, it could have been possible that most or all of the missing data corresponded to orders shipped to customers in a certain state or sales data from a specific week. However, there was no clear pattern or value(s) that corresponded to a majority of the missing 'revenue' values. As such, another method needs to be used to determine how the missing values should be handled.

By plotting the distribution of a numerical variable, it can be used to determine what metrics may be more appropriate for imputing missing values. If a variable is distributed normally (bell-curved), then the mean can be used for imputation; however, if the data is skewed significantly enough, then the mean no longer becomes suitable. Instead, the median would be a more viable statistic for imputation.

```python
#Evaluate the 'revenue' variable to determine how to handle the missing values. Plot the distribution
ax1 = sns.histplot(data=orig_df, x='revenue', kde=True)
#add the median, mean
ax1.axvline(x=orig_df['revenue'].median(), color='orange', linestyle='--', label='Median')
ax1.axvline(x=orig_df['revenue'].mean(), color='green', linestyle='--', label='Mean')
#edit plot details
plt.title('Distribution of Revenue (Last Six Weeks)')
plt.xlabel('Revenue')
plt.ylabel('# Customers / Data Points')
plt.legend(loc='upper right', fontsize=9)
plt.tight_layout();
```

The histogram above illustrates the distribution of the 'revenue' variable. There are about 13,900 data points making up this plot; ~1,100 missing values.  
The distribution appears to be multimodal with a sizable skew to the right. Given that the distribution is far from normal, the median would be more appropriate to impute the missing 'revenue' values with rather than the mean.
- For reference, the median 'revenue' value is 89.5, whereas the average 'revenue' is about 93.9. The former is already at two decimal places so no rounding is necessary.

```python
#Impute missing 'revenue' values with the median
df['revenue'] = df['revenue'].fillna(orig_df['revenue'].median())

#Review the updated dataset
display(df.info())
```

# Analysis IV
**_Based on the data, which method would you recommend we continue to use? Some of these methods take more time from the team so they may not be the best for us to use if the results are similar._**

Thus far, it has been determined that generally, the "Email + Call" sales strategy has yielded the most revenue per customer, whereas only phoning someone has typically produced the least. However, with only a limited number of hours available each week to advertise, it's important that _Pens & Printers_ use their time as effectively as possible. Rather than committing all of the sales team's resources to a single sales technique, a combination of the three methods might yield the most dividends. The obvious question then becomes, how can the team spend their time to optimize sales revenue?

To evaluate this question, some new metrics can be established using the data & information provided. One such metric could quantify the effectiveness or efficiency of a sales method based on the total revenue obtained & the effort/time/cost utilized on a customer.  
Recall that the sales team estimated the amount of time they spend talking to customers on the phone when utilizing the "Call" & "Email + Call" sales techniques. When they only called a customer, the phone call lasted about 30 minutes on average. When they sent an email a week before calling a customer, the phone call lasted about ten minutes per customer. Say that writing & sending an email to a customer takes no longer than three minutes. If so, the typical amount of time per sales method follows as: "Email" (~ 3 minutes), "Email + Call" (~ 13 minutes across one week), & "Call" (~ 30 minutes).
- This in conjunction with the results from **Analysis III** emphasizes how only calling customers (the "Call" method) is the least effective strategy in terms of accruing revenue.

With the usual amount of time per sales method now defined, the relative revenue efficiency per advertisement/customer can be calculated. To do so, the ratio of the typical revenue per customer & the typical amount of time spent on each customer can be obtained for each sales method. For example, the typical revenue per customer in the "Email" group divided by the typical time spent advertising to a customer in this group--about three minutes--gives an estimate of the typical revenue generated per minute. Moreover, this metric can be converted to other time units (i.e. hours, weeks) to generalize.
- As done previously, the median revenue can be used rather than the mean for each sales method since this statistic provides a better assessment of the "typical" revenue. The weekly, median revenues per sales method can be retrieved via the `med_rev_wk_method` variable.
- Moreover, this metric can be evaluated over time, similarly to **Analysis III**, to see how it has evolved since the new line of products was launched six weeks ago.
- Although the units of the 'revenue' were not provided, they are presumed to be US dollars ($).

This metric can be of assistance when evaluating which sales method(s), or some combination of them, yielded the greatest revenue based on the amount of time spent advertising. By using the most recent data, it can provide an assessment of how the company is currently performing in terms of revenue efficiency & how the strategy of the sales team could be altered to improve this performance.

```python
#Define typical time spent per customer by 'sales_method' (minutes)
avg_time_call = 30
avg_time_em = 3
avg_time_em_call = 13

#Calculate revenue efficiency on a weekly basis per 'sales_method'
rev_eff_wk_method = med_rev_wk_method.reset_index(drop=True).copy()
for i in rev_eff_wk_method.index:
    #CALL group
    if rev_eff_wk_method.loc[i,'sales_method'] == 'Call':
        #Revenue efficiency per minute
        rev_eff_wk_method.loc[i,'rev_eff ($/min)'] = np.round(rev_eff_wk_method.loc[i,'med_rev'] / avg_time_call, 3)
        #Revenue efficiency per hour
        rev_eff_wk_method.loc[i,'rev_eff ($/hr)'] = np.round(60*rev_eff_wk_method.loc[i,'med_rev'] / avg_time_call, 3)
        
    #EMAIL group
    elif rev_eff_wk_method.loc[i,'sales_method'] == 'Email':
        #Revenue efficiency per minute
        rev_eff_wk_method.loc[i,'rev_eff ($/min)'] = np.round(rev_eff_wk_method.loc[i,'med_rev'] / avg_time_em, 3)
        #Revenue efficiency per hour
        rev_eff_wk_method.loc[i,'rev_eff ($/hr)'] = np.round(60*rev_eff_wk_method.loc[i,'med_rev'] / avg_time_em, 3)
        
    #EMAIL+CALL group
    else:
        #Revenue efficiency per minute
        rev_eff_wk_method.loc[i,'rev_eff ($/min)'] = np.round(rev_eff_wk_method.loc[i,'med_rev'] / avg_time_em_call, 3)
        #Revenue efficiency per hour
        rev_eff_wk_method.loc[i,'rev_eff ($/hr)'] = np.round(60*rev_eff_wk_method.loc[i,'med_rev'] / avg_time_em_call, 3)
```

```python
#Visualize the revenue-efficiency metric
ax1 = sns.lineplot(data=rev_eff_wk_method, x='week', y='rev_eff ($/hr)', hue='sales_method', 
                   hue_order=['Call','Email','Email + Call'], marker='o')
#edit plot details
plt.title('Typical Weekly Revenue Efficiency per Sales Method (Last Six Weeks)')
plt.xlabel('Week of Sale Since Product Launch')
plt.ylabel('Typical Revenue Efficiency ($/hr)')
plt.yticks(ticks=range(0,2501,250), fontsize=9)
plt.legend(title='Sales Method', title_fontsize=9, fontsize=8)
plt.grid(axis='y',alpha=0.2,color='gray')
plt.tight_layout();
```

The figure above quantifies the typical weekly revenue efficiency that quantifies the usual revenue obtained per customer according to the sales method used. The data spans the last six weeks--since the launch of the new line of office stationery. The efficiency metric measures the typical weekly revenue obtained per customer based on how much time was usually spent advertising to said customer according to the sales method used. This metric is in units of US dollars per hour. The plot was built from 15,000 data points.  
Recall that about half of the data points correspond to customers who were contacted via email, whereas about one-third of the customers were contacted via phone, & about one-sixth were contacted via both methods. Note that each sales method took the sales team differing amounts of time which are listed below.
- _Phone Call only_: Team members were on the phone for around 30 minutes on average per customer.
- _Email only_: It is estimated that constructing & sending an email took no more than three minutes on average.
- _Email + Call_: Customers were sent an email & called a week later. These phone calls were around ten minutes on average per customer.

In **Analysis III**, it was found that per customer, the "Email + Call" strategy yielded the most revenue, whereas the "Call" method yielded the least. When the amount of time (or effort) is taken into account for each of the three sales methods, the picture changes slightly. While the "Call" technique remained the least profitable, the "Email" strategy became the predominant one in this metric. So, in terms of the typical revenue obtained per customer, emailing customers was the most efficient use of the sales team's time, whereas only calling customers was the least efficient.
- Of course, this is assuming that emails take about three minutes to construct & send per customer. The longer this process, the smaller this revenue efficiency becomes for this sales strategy.

Furthermore, the typical revenue metric has generally increased across the past six weeks since the products' launch. Between the first & most recent weeks, the typical revenue efficiencies of each sales method is as follows.
