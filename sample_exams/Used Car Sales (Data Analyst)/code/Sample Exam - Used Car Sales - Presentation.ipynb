## An overview of the project & business goals

#### Intro
In the last six months, _Discount Motors_ has sold nearly 6,750 used cars that span 18 unique models, use various forms of energy for fuel, & are between five & 27 years old as of today (October, 2025). With the company's progression in recent years, _Discount Motors_ has seen a great increase in sales & the number of customers. It is fair to say that the company has grown well despite the ever-changing automotive market.  
Nevertheless, the UK government has recently anncounced that from 2030, all new cars will required to produce zero emissions. As such, it is likely that customers will give more consideration to the future value of their cars. Moreover, _Discount Motors_ will only be able to sell used electric vehicles come 2030; however, none have been sold by the company within the last year.

With this in mind, it poses a sizable challenge for the company to successfully adapt to these potentially significant changes in the automotive market while remaining competitive & profitable. Fortunately, our teams have been working effectively together & have begun to lay out a plan to accomplish this.

#### Business Goals
During the planning process, the following objectives have been established.
- Given that the company has no information regarding recent sales of used electric vehicles, hybrid vehicles can act as a sort of segue between our current situation & the one on the horizon in 2030. Furthermore, by **evaluating how the sales of hybrid vehicles have differed to other types in recent months**, it can provide some direction as to how we can begin promoting & selling used electric vehicles.
- Generally, hybrid & electric vehicles are likely to be more expensive than those that do not use electricity as a source of fuel (diesel & petrol cars). As such, it is crucial that we **increase their appeal** to the average customer.
- Ultimately, the goal is to help the company to **sell more hybrid & electric vehicles in the next year**.

#### Project Overview
Using the last six months of sales data, various characteristics of used cars were explored to find key differences in which hybrid vehicles stood out against others that were sold.

## Your key findings including the metric to monitor & current estimation

#### Findings
One of the first parts of the analytical process involved evaluating how many of the cars sold by _Discount Motors_ were hybrids or not. The majority of the 6,738 cars sold were petrol-based cars (about 60%), which was double the number of hybrid cars sold which can be seen in the figure below. Moreover, these petrol-based vehicles only accounted for about five percent more of the company's total sales than hybrids did. This indicates that hybrid vehicles were more expensive than petrol-based cars on average.

Given these proportions, for the company to sell more hybrid & electric vehicles, it might require sacrificing the number of petrol-based cars that they obtain & sell. If we assume that hybrid & electric vehicles are generally more difficult to sell (at least currently), then the company's total sales might decrease somewhat if less petrol-based vehicles are sold. So, if a dedicated shift is made towards trying to sell more hybrid & electric vehicles & less petrol-based vehicles, sales numbers may decline slightly from what they've been in the last six months.
- Note that over this time period, 6,378 used cars have been sold which have totaled about £84,375,000 in sales.



Throughout the analytical process, one variable in particular stood out when it came to differentiating between hybrid & non-hybrid vehicles; used cars' road taxes. In the last six months, hybrid vehicles had a typical road tax of £20, whereas non-hybrids had a typical road tax of about £145. Moreover, the range of road taxes of hybrid vehicles was noticeably lower than they were for the other types of cars. The figure below illustrates these distributions.  
Given that electric vehicles will have no road tax, advertising this variable will be crucial in the marketing process for both hybrid & electric vehicles.

## Your recommendations to the business

In summary, the main recommendations for the marketing team include the following.
- Explore & research the primary differences between hybrids & electric vehicles versus diesel & petrol vehicles.
    - It is simple to set targets regarding the number of hybrids & electric vehicles to try to sell in the coming years, but executing this process effectively will likely be more challenging due to how many factors there are at play, particularly those that are beyond the company's direct control. With how many used, petrol-based vechicles that _Discount Motors_ has sold recently, it is likely that a sizable customer base depends on this company for such cars. Therefore, the bigger challenge lies in persuading these people to invest in or at least begin to consider more environmentally-friendly vehicles like hybrids & electric vehicles.
    - By researching the primary differences between hybrids & electric vehicles compared to non-environmentally-friendly vehicles (i.e. diesel-based, petrol-based cars), _Discount Motors_ will not only be more knowledgeable in understanding the pros & cons of each group, but they will also be better prepared to answer questions that customers may have about them. Moreover, with the new legislation established by the UK regarding vehicles & their emissions, which is set to come into effect in 2030, the automotive market could see a significant shift in what kinds of cars are used & how they are valued by the general population. Therefore, having a greater understanding of these vehicles will only be of benefit to the company.

- Advertise hybrids & electric vehicles more than any other kind of car. In particular:
    - Emphasize the eco-friendly nature of these cars compared to diesel & petrol cars. Selling cars on the premise of being "eco-friendly" relies on there being customers that are aware of how their car impacts the environment as well as them being interested in minimizing their ecological footprint.
    - Emphasize the lower road taxes & greater fuel efficiencies of these cars compared to diesel & petrol cars. Moreover, the company can point out the stability of fuel consumption metrics of younger, hybrid models compared to those of similarly aged petrol & diesel vehicles.
    - Prioritize used cars that are not too young or too old (the "Goldilocks range"). For hybrids, cars registered 3-5 years prior are likely the most optimal. For electric vehicles, cars registered 2-4 years prior may be the most optimal because younger models should have less technical problems than older ones.
- Slowly shift the inventory of diesel & petrol vehicles so that there are less appealing options to customers.
    - Obtain & house fewer diesel & petrol vehicles with: a small road tax, a low fuel consumption (high efficiency), older models, underpriced cars, etc.
- House & sell more reliable hybrid & electric vehicles at a profitable yet affordable price that is beneficial for both the company & customer. Place greater emphasis on models with less mechanical & technical issues as well as those with a small road tax.
    - Between hybrids & electric vehicles, be more patient & cautious with the latter. Since the company has no recent sales data to go off of, they will be more unpredictable for both the business & customer. Over time, the more electric vehicles that are sold will help paint a clearer picture in terms of how they can be marketed optimally going forward. 
- Monitor the metrics established previously; proportions of hybrids & electric vehicles sold versus diesel & petrol cars.
- Improve the data collection method to allow for more efficient analyses.
    - For example, there were about 100 used cars sold with the fuel type "Other" in the last six months, but the meaning of this fuel type is unclear.
    - More variables can be created & tracked to analyze more characteristics about used cars & their potential relationships with other variables.
- Remain flexible, but be prepared to make changes if necessary. Also, monitor the progression of the 2030 mandate issued by the UK (regarding cars & their emissions) along with similar legislative measures pertaining to the automotive market.
